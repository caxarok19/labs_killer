```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from scipy.optimize import curve_fit
import csv

x = []
y = []

with open('C:\lab2\lab2\lab2\data3-1.txt') as f:
    data = [float(x) for x in (' ').join(f.readlines()).split()]
 
    for i in range(len(data)):
        if i%2==0:
            x.append(data[i])
        else:
            y.append(data[i])

plt.plot(x,y, color = 'red')
plt.title('График зависимости ошибки интерполяции от длины отрезке для N = 3')
plt.xlabel('Длина шага')
plt.ylabel('Ошибка интерполяции ')
plt.legend(['N = 3  для Чебышевского расположения узлов'])
plt.grid()
plt.savefig('data3-1.png')
plt.show()
```
