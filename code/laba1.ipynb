```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from scipy.optimize import curve_fit
import csv

x = []
y = []

with open('C:/Users/PC/Downloads/data3.txt') as f:
    data = [float(x) for x in (' ').join(f.readlines()).split()]
 
    for i in range(len(data)):
        if i%2==0:
            x.append(data[i])
        else:
            y.append(data[i])

plt.plot(x,y, color = 'red')
plt.title('График зависимости логарифма ошибки от логарифма шага для N = 3')
plt.xlabel('Логарифм шага')
plt.ylabel('Логарифм ошибки')
plt.legend(['N = 3'])
plt.grid()
plt.savefig('data3.png')
plt.show()
```

```python
proiz = []
for i in range(len(x)-1):
    proiz.append((y[i+1]-y[i])/(x[i+1]-x[i]))

proiz.reverse()

for i in range(len(proiz)-1):
    if (proiz[i+1]-proiz[i])/(x[i+1]-x[i]) > 1 or (proiz[i+1]-proiz[i])/(x[i+1]-x[i]) < -1:
        print(proiz[i])
        print(i)
        break
 

```
