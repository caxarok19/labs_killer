```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from scipy.optimize import curve_fit
import csv
import math
 
        
x = []
y = []

with open('C:\lab3\lab3\lab3\data.txt') as f:
    data = [float(x) for x in (' ').join(f.readlines()).split()]
 
    for i in range(len(data)):
        if i%2==0:
            x.append(data[i])
        else:
            y.append(data[i])



 
 
plt.plot(x,y, color = 'red')
plt.title('График зависимости логарифма ошибки интерполяции от логарифма количества узлов N')
plt.xlabel('Количество точек')
plt.ylabel('Ошибка интерполяции')
#plt.legend(['N = 3  для Чебышевского расположения узлов'])
plt.grid()
plt.savefig('data.png')
plt.show()
```
